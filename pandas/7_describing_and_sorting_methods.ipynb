# Describing and Sorting Methods:

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('tips.csv')
df
```

## Sorting Values:

```python
df.sort_values('total_bill')
```

```python
df.sort_values('total_bill',ascending=False)
```

### Sorting by considering multiple columns:

```python
df.sort_values(['total_bill','tip'])
```

## Finding Maximum value of an attribute:

```python
df['total_bill'].max()
```

```python
df['total_bill'].idxmax()
```

## Finding Minimum value of an attribute:

```python
df['total_bill'].min()
```

```python
df['total_bill'].idxmin()
```

## Exploring Coorelations among the columns:

```python
df.corr()
```

## Category count method:

### Listing all the unique elements in a column and their freequency of occurence:

```python
df['sex'].value_counts()
```

### Listing of unique values in a column:

```python
df['day'].unique()
```

### Fetching number of uniques values in a column:

```python
df['day'].nunique()
```

## Replace Method(s):

```python
df['sex'].replace(to_replace='Female',value='F')
```

### Replacing multiple values in a column:

```python
df['sex'].replace(['Male','Female','Trans'],['M','F','T'])
```

### Replacing using maps method:

```python
mymaps={'Male':'M','Female':'F','LGBTQ+':'T'}
df['sex'].replace(mymaps)
```

## Working with duplicate rows:

```python
df1=pd.DataFrame(data=[[1,2,3],[4,3,2],[1,2,3],[7,7,8],[9,0,1],[1,2,3],[7,7,8]])
df1
```

#### Detecting Duplicates:

```python
df1.duplicated()
```

#### Dropping Duplicates:

```python
df1.drop_duplicates()
```

## Filtering rows based on values:

```python
df['total_bill'].between(10,20,inclusive=True)
```

## Grabbing n largest column values:

```python
df.nlargest(10,'tip')
```

## Grabbing n smallest column values:

```python
df.nsmallest(10,'tip')
```

## Fetching a sample data:

```python
df.sample(20)
```
