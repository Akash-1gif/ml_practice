# Conditional Filtering:

```python
import numpy as np
import pandas as pd
```

```python
data1=[[1776,328,20.5],[1867,38,1.7],[1821,126,1.22]]
column1=['Year','Pop','GDP']
index1=['USA','Canada','Mexico']
df=pd.DataFrame(data=data1,columns=column1,index=index1)
df
```

## Query1: Pop>50:

```python
df[df['Pop']>50]
```

```python
(df[df['Pop']>50]).index
```

```python
(df[df['Pop']>50])['GDP']
```

```python
df=pd.read_csv('tips.csv')
df
```

## Single Condition Filtering:

```python
x1=df[df['sex']=='Female']
x1
```

```python
x2=df[df['total_bill']>=40]
x2
```

## Multiple Conditions Filtering:

```python
x3=df[(df['total_bill']>=40) & (df['sex']=='Female')]
x3
```

```python
x4=df[(df['sex']=='Female') | (df['total_bill']>=40)]
x4
```

```python
x5=df[((df['sex']=='Male') & (df['total_bill']>=40)) & ((df['day']=='Sat') | (df['day']=='Sun'))]
x5
```

## Multiple Conditions Filtering:

```python
df['total_bill'].apply(lambda num: num*2)
```
