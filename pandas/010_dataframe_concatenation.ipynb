# Combining/Concatenating DataFrames:

```python
import numpy as np
import pandas as pd
```

```python
d1={'A':['A0','A1','A2','A3'],'B':['B0','B1','B2','B3']}
d2={'C':['C0','C1','C2','C3'],'D':['D0','D1','D2','D3']}
d1=pd.DataFrame(d1)
d2=pd.DataFrame(d2)
d1
```

```python
d2
```

## Combining DataFrames along the columns:

```python
pd.concat([d1,d2],axis=1)
```

## Combining DataFrames along the rows:

```python
pd.concat([d1,d2],axis=0)
```

```python
d2.columns=d1.columns
d2
```

```python
d3=pd.concat([d1,d2],axis=0)
d3
```

```python
d3.index=range(len(d3))
d3
```
