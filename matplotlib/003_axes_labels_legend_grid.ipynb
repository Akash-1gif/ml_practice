# Axes, Labels, Legend, Grid

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
plt.style.use('dark_background')
```

```python
x=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

### Creating graphs based on the given data:

```python
plt.plot(x,max_t)
plt.plot(x,min_t)
plt.plot(x,avg_t)
```

### Labelling the generated graph:

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title("Weather")
plt.plot(x,max_t,label='Max Temp')
plt.plot(x,min_t,label='Min Temp')
plt.plot(x,avg_t,label='Avg Temp')
plt.legend()
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title("Weather")
plt.plot(x,max_t,label='Max Temp')
plt.plot(x,min_t,label='Min Temp')
plt.plot(x,avg_t,label='Avg Temp')
plt.legend(loc="upper right")
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title("Weather")
plt.plot(x,max_t,label='Max Temp')
plt.plot(x,min_t,label='Min Temp')
plt.plot(x,avg_t,label='Avg Temp')
plt.legend(loc="best")
```

## Grid:

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title("Weather")
plt.plot(x,max_t,label='Max Temp',color='red',linewidth=3,marker='o',markersize=6,markerfacecolor='yellow')
plt.plot(x,min_t,label='Min Temp',color='green',linewidth=3,marker='o',markersize=6,markerfacecolor='yellow')
plt.plot(x,avg_t,label='Avg Temp',color='blue',linewidth=3,marker='o',markersize=6,markerfacecolor='yellow')
plt.legend(loc="best")
plt.grid()
```
