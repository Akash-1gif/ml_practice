# Bar Graph:

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
plt.style.use('dark_background')
```

```python
x=['GOOGLE','AMAZON','META','APPLE']
y=[320,280,300,250]
z=[80,120,60,75]
xlot=np.arange(len(x))
xlot
```

## Generating a bar graph:

```python
plt.xticks(xlot,x)
plt.bar(xlot,y,color='red')
plt.xlabel('Company')
plt.ylabel('Revenue')
plt.title('Yearly Revenue:')
plt.grid()
plt.show()
```

```python
plt.xticks(xlot,x)
plt.bar(xlot,y,color='red',label='Revenue')
plt.bar(xlot,z,color='green',label='Profit')
plt.xlabel('Company')
plt.ylabel('Billions')
plt.title('Revenue and Profits:')
plt.legend(loc='best')
plt.grid()
plt.show()
```

```python
plt.xticks(xlot,x)
plt.bar(xlot-0.2,y,color='red',label='Revenue',width=0.4)
plt.bar(xlot+0.2,z,color='green',label='Profit',width=0.4)
plt.xlabel('Company')
plt.ylabel('Billions')
plt.title('Revenue and Profits:')
plt.legend(loc='best')
plt.grid()
plt.show()
```

## Horizontal Bar Graphs:

```python
plt.yticks(xlot,x)
plt.barh(xlot-0.2,y,color='red',label='Revenue',height=0.4)
plt.barh(xlot+0.2,z,color='green',label='Profit',height=0.4)
plt.ylabel('Company')
plt.xlabel('Billions')
plt.title('Revenue and Profits:')
plt.legend(loc='best')
plt.grid()
plt.show()
```
