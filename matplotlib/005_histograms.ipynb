# Histogram

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('dark_background')
```

```python
team1_speed=[250,270,215,222,300,253,212,217,290,274,244,230,283,216,232,271,245,263,296,267,269]
team2_speed=[212,246,232,240,289,290,298,213,210,245,255,263,271,222,234,262,253,221,289,215,221]
```

### Plotting a basic histogram:

```python
plt.hist(team1_speed)
plt.xlabel('Top Speed')
plt.ylabel('No of Cars')
```

### Setting up bins range:

```python
plt.hist(team1_speed,bins=[200,250,280,300])
plt.xlabel('Top Speed')
plt.ylabel('No of Cars')
```

### Setting up bar width:

```python
plt.hist(team1_speed,bins=[200,230,260,290,320],rwidth=0.95)
plt.xlabel('Top Speed')
plt.ylabel('No of Cars')
```

### Displaying steps instead of bars:

```python
plt.hist(team1_speed,bins=[200,230,260,290,320],rwidth=0.95,histtype='step')
plt.xlabel('Top Speed')
plt.ylabel('No of Cars')
```

### Plotting histogram for multiple datasets:

```python
plt.hist([team1_speed,team2_speed],bins=[200,230,260,290,320],rwidth=0.90,color=['orange','green'],label=['Team Splitfire','Team Horizon'])
plt.xlabel('Top Speed')
plt.ylabel('No of Cars')
plt.legend()
```

### PLotting Horizontal histogram:

```python
plt.hist([team1_speed,team2_speed],bins=[200,230,260,290,320],rwidth=0.90,color=['orange','green'],label=['Team Splitfire','Team Horizon'],orientation='horizontal')
plt.xlabel('Top Speed')
plt.ylabel('No of Cars')
plt.legend(loc='best')
```

# Conclusion:
### * Team Splitfire has more number of Hypercars and Present generation's supercars
### * Team Horizon has more number of Basic Supercars and past generation's Hypercars
