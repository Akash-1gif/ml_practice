# Pie Charts:

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('dark_background')
```

### Data:

```python
expenses=[3649,1500,109,435,300]
expense_label=['Rent','Monthly Expenses','Phone/Internet bill','Bus Pass','Miscellaneous'] #damn, this was expensive... :(
```

### Pie Chart:

```python
plt.pie(expenses,labels=expense_label)
plt.show()
```

### Making a proper cirle:

```python
plt.axis('equal')
plt.pie(expenses,labels=expense_label)
plt.show()
```

## Adding percentage value:

```python
plt.pie(expenses,labels=expense_label,autopct='%0.2f%%')
plt.show()
```

### Adding colours:

```python
colors=['red','green','blue','yellow','orange']
plt.pie(expenses,labels=expense_label,autopct='%0.2f%%',colors=colors)
plt.axis('equal')
plt.show()
```

```python
colors=['red','green','blue','yellow','orange']
plt.pie(expenses,labels=expense_label,autopct='%0.2f%%',colors=colors)
plt.axis('equal')
plt.legend(loc='best')
plt.show()
```

### Using explode function:

```python
colors=['red','green','blue','yellow','orange']
plt.pie(expenses,labels=expense_label,autopct='%0.2f%%',colors=colors,explode=[0,0.3,0,0,0])
plt.axis('equal')
plt.savefig("images/piechart.jpg",bbox_inches="tight",pad_inches=2)
plt.show()
```
