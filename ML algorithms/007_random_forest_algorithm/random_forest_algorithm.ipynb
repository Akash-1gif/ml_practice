# Random Forest Algorithm:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_digits
from sklearn.ensemble import RandomForestClassifier
%matplotlib inline
plt.style.use('dark_background')
```

```python
digits=load_digits()
dir(digits)
```

```python
for i in range(4):
    plt.matshow(digits.images[i])
```

## Preparing data:

```python
x=pd.DataFrame(data=digits.data)
x
```

```python
y=pd.DataFrame(data=digits.target)
y
```

## Training out prediction model:

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
len(x_train)
```

```python
len(y_test)
```

```python
model=RandomForestClassifier()
```

```python
model.fit(x_train,y_train[0])
```

```python
model.score(x_test,y_test)
```

```python
accuracy=[]
```

```python
p=np.arange(10,501,10)
p
```

```python
for i in p:
    model=RandomForestClassifier(n_estimators=i)
    model.fit(x_train,y_train[0])
    q=model.score(x_test,y_test)*100
    accuracy.append(q)
    
accuracy
```

```python
plt.scatter(p,accuracy,color='green')
plt.plot(p,accuracy,color='red')
plt.xlabel('No of Randomm Forests')
plt.ylabel('Accuracy')
plt.show()
```

```python
df=pd.DataFrame(data=p,columns=['No of Decision Tree Tables'])
df
```

```python
df['Prediction Accuracy']=accuracy
```

```python
df
```

```python
df.corr()
```

```python
df["Prediction Accuracy"].value_counts()
```

```python
df["Prediction Accuracy"].max()
```
