# Titanic Survival Rate Prediction:

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

### Importing data:

```python
df=pd.read_csv('titanic.csv')
df
```

### Preparing data:

```python
df['Pclass'].value_counts()
```

```python
df['Pclass'].isnull().sum()
```

```python
df['Sex'].isnull().sum()
```

```python
df['Age'].isnull().sum()
```

```python
df['Fare'].isnull().sum()
```

```python
df_dead=df[df['Survived']==0]
df_dead
```

```python
df_dead_male=df_dead[df_dead['Sex']=="male"]
df_dead_male
```

```python
df_dead_male['Age']=df_dead_male['Age'].fillna(df_dead_male['Age'].mean())
df_dead_male
```

```python
df_dead_male['Age'].isnull().sum()
```

```python
df_dead_female=df_dead[df_dead['Sex']=="female"]
df_dead_female
```

```python
df_dead_female['Age']=df_dead_female['Age'].fillna(df_dead_female['Age'].mean())
df_dead_female
```

```python
df_dead_female['Age'].isnull().sum()
```

```python
df_dead=pd.concat([df_dead_male,df_dead_female],axis=0)
df_dead
```

```python
df_dead['Age'].isnull().sum()
```

```python
df_alive=df[df['Survived']==1]
df_alive
```

```python
df_alive_male=df_alive[df_alive['Sex']=='male']
df_alive_male
```

```python
df_alive_male['Age']=df_alive_male['Age'].fillna(df_alive_male['Age'].mean())
df_alive_male
```

```python
df_alive_male['Age'].isnull().sum()
```

```python
df_alive_female=df_alive[df_alive['Sex']=="female"]
df_alive_female
```

```python
df_alive_female['Age']=df_alive_female["Age"].fillna(df_alive_female['Age'].mean())
df_alive_female
```

```python
df_alive_female["Age"].isnull().sum()
```

```python
df_alive=pd.concat([df_alive_male,df_alive_female],axis=0)
df_alive
```

```python
df_alive['Age'].isnull().sum()
```

```python
df=pd.concat([df_alive,df_dead],axis=0)
```

```python
df
```

```python
df['Age'].isnull().sum()
```

```python
target=df['Survived']
target
```

```python
inputs=df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis=1)
inputs
```

```python
le_S=LabelEncoder()
```

```python
inputs['Sex_l']=le_S.fit_transform(inputs['Sex'])
inputs
```

```python
inputs=inputs.drop(['Sex'],axis=1)
inputs
```

## Training Data:

```python
model=DecisionTreeClassifier()
```

```python
x_train,x_test,y_train,y_test=train_test_split(inputs,target,test_size=0.2,random_state=11)
```

```python
x_train
```

```python
y_train
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
