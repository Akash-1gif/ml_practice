# Implementing Linear Regression module to Predict the Property Prices:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import linear_model
plt.style.use('dark_background')
```

### Importing the training data:

```python
df=pd.read_csv('homeprices.csv')
df
```

```python
plt.xlabel('Area (in sq.ft)') #Independent variable
plt.ylabel('Price (in USD)') #Dependent Variable
plt.scatter(df['area'],df['price'],label='Training Data')
plt.legend()
plt.show()
```

### Creating a linear regression object model:

```python
reg=linear_model.LinearRegression()
```

### Training the linear regression object model with the training data:

```python
reg.fit(df[['area']],df['price']) 
```

```python
plt.xlabel('Area (in sq.ft)') #Independent variable
plt.ylabel('Price (in USD)') #Dependent Variable
plt.scatter(df['area'],df['price'],label='Training data')
plt.plot(df['area'],reg.predict(df[['area']]),label='Predicted values',color='orange')
plt.legend(loc="best")
plt.show()
```

### Finding the calculated coefficient and intercept:

```python
reg.coef_
```

```python
reg.intercept_
```

### Predicting the Price of the 3300sq ft area:

```python
reg.predict(pd.DataFrame(data=[[3300]]))
```
