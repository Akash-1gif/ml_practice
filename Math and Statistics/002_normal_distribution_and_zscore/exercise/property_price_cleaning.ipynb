```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('bhp.csv')
df
```

```python
df1=df.copy()
df1
```

## Setting up limits between 99.9 and 0.1 quantile:

```python
l1=df1['price_per_sqft'].quantile(0.001)
l1
```

```python
l2=df1['price_per_sqft'].quantile(0.999)
l2
```

```python
df1=df1[(df1['price_per_sqft']<=l2)&(df1['price_per_sqft']>=l1)]
df1
```

```python
df2=df1.copy()
df2
```

```python
sns.histplot(df1['price_per_sqft'],kde=True)
```

```python
mean_x=df1['price_per_sqft'].mean()
mean_x
```

```python
std_x=df1['price_per_sqft'].std()
std_x
```

```python
l1=mean_x+(std_x*3)
l1
```

```python
l2=mean_x-(std_x*3)
l2
```

```python
df2=df2[(df2['price_per_sqft']>=l2)&(df2['price_per_sqft']<=l1)]
df2.head()
```

```python
df2.describe()
```

## Calculating z-score for each row on the basis of price_per_sqft:

```python
df3=df1.copy()
df3.describe()
```

```python
df3['z-score']=(df3['price_per_sqft']-mean_x)/std_x
df3.sample(10)
```

```python
df3[(df3['z-score']>=3)|(df3['z-score']<=(-3))]
```

```python
df3=df3[(df3['z-score']<=3)&(df3['z-score']>=(-3))]
df3.sample(10)
```

```python
df3.describe()
```
